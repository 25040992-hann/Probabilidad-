```python
from statistics import *
```

```python
# lista de calificaciones del grupo de probabilidad y estadística
grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
type(grupoa)
```

```python
# Sacar la media de las calificaciones con una formula
promedio = sum(grupoa)/len(grupoa)
print(promedio)
```

```python
# Utilizar la funcion mean que esta en la libreria statistics para sacar el promedio
promedio2 = mean(grupoa)
print(promedio2)
```

```python
# Detreminar el valor de la meidana con la funcion median
mediana = median(grupoa)
print(mediana)
```

```python
# Determinar el valor de la moda con la funcion mode
moda = mode(grupoa)
print(moda)
```

```python
# En caso de que exista mas de una moda, se puede emplear la funcion multimode
# para sacar multiples modas que puedan existir
moda2 = multimode(grupoa)
print(moda2)
```

```python
# Definir una variablellamada tiempos100mts de tipo lista, que contiene los
# tiempos que tardaron los 10 alumnos para corre los 100 metros
tiempos100mts = [15.10, 17.20, 14.69, 13.27, 22.15, 18.71, 19.15, 20.65, 15.20,
                 17.28]
print(tiempos100mts)
```

```python
# Usamos la funcion mean para determinar el tiempo promedio de la lista de valores
# Usar la funcion median para determinar la mediana de los tiempos
# Usar la funcion mode para determinar la moda de la lista de valores
print("el tiempo promedio de los corredores es de: ", mean(tiempos100mts))
print("la media del tiempo es de: ", median(tiempos100mts))
print("la moda de la lista de valores es: ", mode(tiempos100mts))
```

```python
# Libreria para generar numeros aleatorios
import random
```

```python
fabrica1  = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50, 100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2  = []
for i in range(100):
  numero_aleatorio2 = round(random.uniform(70, 100), 2)
  fabrica2.append(numero_aleatorio2)
print(fabrica2)
```

```python
# Libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("producción diaria de la fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("producción diaria de la fabrica 2")
```

```python
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("VALORES DE LA FABRICA 1")
print("la media es = ", media1)
print("la mediana es = ", mediana1)
print("la moda es = ", moda1)
```

```python
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print("VALORES DE LA FABRICA 2")
print("la media es = ", media2)
print("la mediana es = ", mediana2)
print("la moda es = ", moda2)
```
