###Problema 1 - Poisson

```python
from scipy.stats import poisson
```

```python
media_personas_por_dia = 5
probabilidad_2_personas = poisson.pmf(2, media_personas_por_dia)
print("1. La probabilidad de que el número de personas sean ingresadas a una unidad de cuidados intensivos:",
      probabilidad_2_personas)

x = 10
probabilidad_10_o_mas = poisson.pmf(x, media_personas_por_dia)

print("2. La probabilidad de que se exceda de 10 es: ",
      probabilidad_10_o_mas)
```

###Problema 2 - Binomial

```python
from scipy.stats import binom
```

```python
probabilidad_de_defectos = 0.001
lotes = 1000
X = 1000

probabilidad_ninguno_sea_defectuoso = binom.pmf(X, lotes, probabilidad_de_defectos)
print("La probabilidad de que ningúno sea defectuoso es: ", probabilidad_ninguno_sea_defectuoso)

probabilidad_3_defectuosos = binom.pmf(3, lotes, probabilidad_de_defectos)
print("La probabilidad de que exactamente 3 sean defectuosos es: ", probabilidad_3_defectuosos)

probabilidad_4_defectuosos = binom.pmf(4, lotes, probabilidad_de_defectos)
```
