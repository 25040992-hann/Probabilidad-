```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = [0, 1, 2, 3, 4, 5]
p = [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]

print("Suma de probabilidades: ", sum(p))

plt.stem(x, p)
plt.title("Distribución de Probabilidad")
plt.xlabel("Num. de Celulas Anormales")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
p = [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]

valor_esperado = 0

for i in range(len(x)):
  xi = x[i]
  pi = p[i]
  producto = xi * pi
  valor_esperado += producto

  print(f"x = {xi}, p = {pi}, x*p = {producto:.2f}")
print(f"\nValor esperado (Esperanza): {valor_esperado:.2f}")
```

```python
x = [0, 1, 2, 3, 4, 5]
p = [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]

esperanza = 0
for i in range(len(x)):
  esperanza+=  x[i] * p[i]

esperanza_cuadrada = 0
for i in  range(len(x)):
  esperanza_cuadrada += (x[i] ** 2) * p[i]

varianza = esperanza_cuadrada - esperanza**2
desviacion = math.sqrt(varianza)

print(f"E(x) = {esperanza:.2f}")
print(f"E(x^2) = {esperanza_cuadrada:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviación Estándar = {desviacion:.2f}")
```

```python
x = [0, 1, 2, 3, 4, 5]
p = [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]

F = np.cumsum(p)

for xi, fi in zip(x, F):
  print(f"F({xi}) = {fi:.2f}")

plt.step(x, F, where='post')
plt.title("Función de Distribución Acumulada (FDA)")
plt.xlabel("Num. de Celulas Aniormales")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```

##Interpretación de Resultados
-> El promedio esperado de las células anormales por muestra es moderado.

-> La variable no se refleja como excesiva, lo que indica que las muestras sulen presentar un numero relativamente estable en las células anormales.

-> La función cumulada muestra que es mas probable encontrar un número bajo o intermedio de células anormales.
