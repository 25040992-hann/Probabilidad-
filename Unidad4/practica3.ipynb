#Distribución Binomial

###Ejercicio 1

```python
from scipy.stats import binom
```

```python
probabilidad_exito = 0.83
num_pacientes = 16
X = 16

probabilidad_ningun_enfermo = binom.pmf(X, num_pacientes, probabilidad_exito)

print("La probabilidad de que ningún pacientes sufra la enfermedad es: ", probabilidad_ningun_enfermo)
```

```python
probabilidad_exito = 0.83
num_pacientes = 16
X = 0

probabilidad_todos_enfermos = binom.pmf(X, num_pacientes, probabilidad_exito)

print("La probabilidad de que todos los pacientes sufran la enfermedad es de: ", probabilidad_todos_enfermos)
```
