```python
from statistics import *
```

```python
# lista de calificaciones del grupo de probabilidad y estadística
grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
type(grupoa)
```

```python
# Sacar la media de las calificaciones con una formula
promedio = sum(grupoa)/len(grupoa)
print(promedio)
```

```python
# Utilizar la funcion mean que esta en la libreria statistics para sacar el promedio
promedio2 = mean(grupoa)
print(promedio2)
```

```python
# Detreminar el valor de la meidana con la funcion median
mediana = median(grupoa)
print(mediana)
```

```python
moda = mode(grupoa)
print(moda)
```

```python
moda2 = multimode(grupoa)
print(moda2)
```
