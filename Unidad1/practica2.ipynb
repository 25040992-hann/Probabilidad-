```python
import random
import matplotlib.pyplot as plt
from statistics import*
```

```python
#generar 100 numeros aleatorios que representan la producción
#en metros de tela de 100 días
fabrica1 = []
for i in range(100):
  numero_aleatorio =round(random.uniform(50, 100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
#generar 100 numeros aleatorios que representan la producción
#en metros de tela de 100 días
fabrica2 = []
for i in range(100):
  numero_aleatorio =round(random.uniform(50, 100), 2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
maximo1 = max(fabrica1)
minimo1 = min(fabrica1)
rango = maximo1 - minimo1
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)
desviacion_estandar_fabrica1 =stdev(fabrica1)
print("estadistica descriptiva de la fabrica 1 ", "media =", media1,
      "mediana =", mediana1, "moda =", moda1, "maximo =", maximo1,
      "minimo =", minimo1, "rango =", rango1)
print(desviacion_estandar_fabrica1)
```
