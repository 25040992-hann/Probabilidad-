```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
#cargar el archivo CSV en una variable llamda covid
autos = pd.read_csv("autos2.csv", encoding = "ISO-8859-1")
```

```python
autos.info()
```

```python
minimoprecios = min(autos['price'])
maximoprecios = max(autos['price'])
print("El minimo de precios es de:", minimoprecios)
print("El maximo de precios es de:", maximoprecios)
```

```python
from statistics import *
import matplotlib.pyplot as plt
```

```python
precios_autos = autos['price'].dropna()

rangoprecios = maximoprecios - minimoprecios
media = mean(precios_autos)
mediana = median(precios_autos)
moda = mode(precios_autos)
desviacion_estandar_autos = stdev(precios_autos)

print("estadistica descriptiva de autos \n", "media =", media,
      "\n mediana =", mediana, "\n moda =", moda, "\n rango =", rangoprecios)
print("desviación estándar =", desviacion_estandar_autos)
```

```python
import matplotlib.pyplot as plt
```

```python
media = autos.price.mean()
plt.hist(autos.price.dropna(), bins = 11)
plt.title("Histograma de precios de autos")
plt.axvline(media, color='y', linestyle='dashed', linewidth=2, label=f'media={media:.2f}')
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.legend()
```

El histograma muestra que la mayoría de los autos se concentran principalmente en un rango de precios bajos, esto indica que el mercado está dominado por vehículos mas económicos.
Conforme el precio aumenta, la frecuencia disminuye reflejando que los autos de lujo son mucho menos comunes, hay muchos valores pequeños y pocos valores muy grandes que alargan la cola.

```python
plt.hist(autos.price.dropna(), bins = 3)
plt.title("Histograma de precios de autos")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
sns.boxplot(x="price", y="body-style", data=autos)
plt.title("Diagrama de precios de autos")
plt.xlabel("Precios")
plt.ylabel("Estilo")
```

El diagrama revela que la mayoría de los precios se concentran en un rango bajo, lo cual se observa en la parte central de la caja.
La mediana se ubica hacia la parte inferior de la caja, lo que confirma que hay muchos autos económicos y pocos autos de precio muy alto.
Los bigotes muestran el rango de precios más común, mientras que los puntos representan autos de lujo o de gama alta que tienen precios mucho mas altos al resto, amplíando la dispersión.



```python
sns.boxplot(x="price", y="drive-wheels", data=autos)
plt.title("Diagrama de tipo de traccion de los autos")
plt.xlabel("Precios")
plt.ylabel("Estilo")
```

El gráfico permite observar cómo se distribuyen los precios según el tipo de tracción, generalmente, los autos con tracción delantera muestran precios más bajos esto lo que hace es reflejar que son los más comunes y económicos en el mercado. En cambio, los autos con tracción trasera y en las ruedas tienden a tener precios más elevados y una mayor dispersión, lo que indica que incluyen modelos de gama media y alta.
Tambien se señala la presencia de autos de lujo cuyo precio se aleja mucho del resto.
Esto ayuda a comprender que el tipo de tracción está relacionado con el segmento del mercado.

```python
sns.boxplot(x="price", y="engine-location", data=autos)
plt.title("Diagrama de ubicacion del motor en los autos")
plt.xlabel("Precios")
plt.ylabel("Estilo")
```

El gráfico muestra cómo varían los precios según la ubicación del motor, los autos con motor delantero concentran la mayoría de los datos, con precios más bajos, lo que refleja que son los más comunes y accesibles en el mercado.
En contraste, los autos con motor trasero presentan precios más elevados, lo que indica que corresponden a modelos de lujo.
Tambien se señala en cada categoría los autos con precios extremos que se alejan del rango típico.
