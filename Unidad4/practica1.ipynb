```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#Generar datos para la distribucion normal
media = 70
desv_std = 4
num_datos = 500

#Generar los datos de maenra aleatoria y ponerlo en una lista
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] #muetra los primeros 50 numeros
```

```python
#Crear un rango de valores para crear la densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

#Calcular la PDF de la distribucion normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 /  (2 * desv_std**2))
print("Imprimir los primeros 50 valores de x", x[0:50], " , y los primeros 50 pdf", pdf[0:50])
```

```python
#Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Norma:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Función de Densiada de Probabilidad (PDF) - Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot(x, pdf, 'k', linewidth=2, )

plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```
