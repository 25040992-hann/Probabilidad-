```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = [0, 10, 20, 50]
p = [0.50, 0.30, 0.15, 0.05]

print("Suma de probabilidades: ", sum(p))

plt.stem(x, p)
plt.title("Distribución de Probabilidad")
plt.xlabel("Boletos")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

```python
x = [0, 10, 20, 50]
p = [0.50, 0.30, 0.15, 0.05]

valor_esperado = 0

for i in range(len(x)):
  xi = x[i]
  pi = p[i]
  producto = xi * pi
  valor_esperado += producto

  print(f"x = {xi}, p = {pi}, x*p = {producto:.2f}")
print(f"\nValor esperado (Esperanza): {valor_esperado:.2f}")
```

```python
x = [0, 10, 20, 50]
p = [0.50, 0.30, 0.15, 0.05]

esperanza = 0
for i in range(len(x)):
  esperanza+=  x[i] * p[i]

esperanza_cuadrada = 0
for i in  range(len(x)):
  esperanza_cuadrada += (x[i] ** 2) * p[i]

varianza = esperanza_cuadrada - esperanza**2
desviacion = math.sqrt(varianza)

print(f"E(x) = {esperanza:.2f}")
print(f"E(x^2) = {esperanza_cuadrada:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviación Estándar = {desviacion:.2f}")
```

```python
x = [0, 10, 20, 50]
p = [0.50, 0.30, 0.15, 0.05]

F = np.cumsum(p)

for xi, fi in zip(x, F):
  print(f"F({xi}) = {fi:.2f}")

plt.step(x, F, where='post')
plt.title("Función de Distribución Acumulada (FDA)")
plt.xlabel("Boletos")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```

##Interpretación de Resultados
-> El valor esperado representa la ganancia promedio que un jugador podría llegar a obtener en el caso de que participara repetidamente, cumulando 8.50 cada vez que vuelve a jugar.

-> El juego no es favorable para el jugador porque aunque hay premios grandes la posibilidad de llegar a obtenerlos es baja y lo que se gasta suponiendo que el boleto cueste mas de 8.50 pesos no es equivalente ni conveniente.
