#Distribución de Poisson

###Ejercicio 1

```python
from scipy.stats import poisson
```

```python
media_accidentes = 2
probabilidad_0_accidentes = poisson.pmf(0, media_accidentes)

print("1. Probailidad de que no haya accidentes en 1 semana: ", probabilidad_0_accidentes)
media_2_semanas = media_accidentes * 2

probabilidad_3_o_menos_accidentes = poisson.cdf(3, media_2_semanas)
print("2. Probabilidad de uqe a lo mucho haya tres accidentes: ", probabilidad_3_o_menos_accidentes)
```
