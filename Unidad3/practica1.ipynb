```python
import matplotlib.pyplot as plt
```

```python
x = [5, 6, 7, 8, 9, 10]
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]

print("Suma de probabilidades: ", sum(p))

plt.stem(x, p)
plt.title("Distribución de Probabilidades")
plt.xlabel("Calificación")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```
